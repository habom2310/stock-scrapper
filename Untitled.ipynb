```python
import bs4
import requests
from tqdm import tqdm
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver = webdriver.Chrome("chromedriver_win32/chromedriver.exe")
```

```python
urls = []
with open("filtered_urls.txt", "r") as text_file:
    urls = text_file.readlines()
```

```python
driver.find_element_by_xpath('//*[contains(@class, "symbol")]')
```

```python
cols = ["Ma ck", "gia", "eps", "pe", "gt so sach", "klgd10", "kl niem yet", "kl luu hanh", 'von hoa']
df_info = pd.DataFrame(columns=cols)
for url in tqdm(urls[1152:]):
    driver.get(url[:-1]) 
    
    try:
        mack = driver.find_element_by_id("symbolbox").text
    except:
        continue
    
    try:
        price = float(driver.find_element_by_id("REF").text)
    except:
        price = None
        continue
        
    try:
        eps_ele = driver.find_elements_by_xpath('//*[contains(@id, "pnEPS")]')[0]
        eps = float(eps_ele.find_element_by_class_name("r").text.replace(",",""))
    except:
        eps = None
        continue

    info_ele = driver.find_elements_by_class_name("dltl-other")
    
    clearfix_ele = []
    
    for v in info_ele:
        clearfix_ele.extend(v.find_elements_by_class_name("clearfix"))
    
    try:
        pe = float(clearfix_ele[2].find_element_by_class_name("r").text.replace(",",""))
    except:
        pe = None
        
    try:
        gtsosach = float(clearfix_ele[3].find_element_by_class_name("r").text.replace(",",""))
    except:
        gtsosach = None
    
    try:
        klgd10 = float(clearfix_ele[5].find_element_by_class_name("r").text.replace(",",""))
    except:
        klgd10 = None
    try:
        klniemyet = float(clearfix_ele[6].find_element_by_class_name("r").text.replace(",",""))
    except:
        klniemyet = None
    try:    
        klluuhanh = float(clearfix_ele[7].find_element_by_class_name("r").text.replace(",",""))
    except:
        klluuhanh = None
    try:    
        vonhoa = float(clearfix_ele[8].find_element_by_class_name("r").text.replace(",",""))
    except:
        vonhoa = None
    
    x = [mack, price, eps, pe, gtsosach, klgd10, klniemyet, klluuhanh, vonhoa]
    df_x = pd.DataFrame([x], columns=cols)
    
    df_info = df_info.append(df_x, ignore_index = True)
```
